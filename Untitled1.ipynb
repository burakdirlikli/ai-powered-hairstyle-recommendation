```python
from PIL import Image

def center_crop_resize_overwrite(path, size=1024):
    img = Image.open(path).convert("RGB")
    w, h = img.size

    crop_size = min(w, h)
    left = (w - crop_size) // 2
    top = (h - crop_size) // 2
    right = left + crop_size
    bottom = top + crop_size

    img = img.crop((left, top, right, bottom))
    img = img.resize((size, size), Image.LANCZOS)
    img.save(path)

    print(f"{path} -> {Image.open(path).size}")

center_crop_resize_overwrite("/content/HairFastGAN/input/user.png")
center_crop_resize_overwrite("/content/HairFastGAN/input/hair_ref.png")
```

```python
from PIL import Image
import os

BASE_DIR = "/content/HairFastGAN"
INPUT_DIR = f"{BASE_DIR}/input"
OUTPUT_DIR = f"{BASE_DIR}/output"

os.makedirs(INPUT_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)

user_file = "user.png"
hair_refs = [
    "hair_ref1.png",
    "hair_ref2.png",
    "hair_ref3.png",
    "hair_ref4.png",
    "hair_ref5.png"
]

def center_crop_resize_overwrite(path, size=1024):
    img = Image.open(path).convert("RGB")
    w, h = img.size

    crop_size = min(w, h)
    left = (w - crop_size) // 2
    top = (h - crop_size) // 2
    right = left + crop_size
    bottom = top + crop_size

    img = img.crop((left, top, right, bottom))
    img = img.resize((size, size), Image.LANCZOS)
    img.save(path)

    print(f"{os.path.basename(path)} -> {Image.open(path).size}")

# user görselini düzenle
center_crop_resize_overwrite(f"{INPUT_DIR}/{user_file}")

# 5 hair ref görselini düzenle
for ref in hair_refs:
    center_crop_resize_overwrite(f"{INPUT_DIR}/{ref}")
```

```python
import os
import subprocess

BASE_DIR = "/content/HairFastGAN"
INPUT_DIR = f"{BASE_DIR}/input"
OUTPUT_DIR = f"{BASE_DIR}/output"

user_file = "user.png"
hair_refs = [
    "hair_ref1.png",
    "hair_ref2.png",
    "hair_ref3.png",
    "hair_ref4.png",
    "hair_ref5.png"
]

os.chdir(BASE_DIR)

result_files = []

for i, ref in enumerate(hair_refs, start=1):
    result_name = f"result_{i}.png"
    result_path = f"{OUTPUT_DIR}/{result_name}"

    cmd = [
        "python", "main.py",
        "--face_path", user_file,
        "--shape_path", ref,
        "--color_path", ref,
        "--input_dir", "input",
        "--result_path", result_path
    ]

    print(f"\nRunning {i}/5 with {ref} ...")
    subprocess.run(cmd, check=True)
    result_files.append(result_name)

print("\nTamamlandı.")
print("Result files:", result_files)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

BASE_DIR = "/content/HairFastGAN"
INPUT_DIR = f"{BASE_DIR}/input"
OUTPUT_DIR = f"{BASE_DIR}/output"

user_file = "user.png"
hair_refs = [
    "hair_ref1.png",
    "hair_ref2.png",
    "hair_ref3.png",
    "hair_ref4.png",
    "hair_ref5.png"
]

result_files = [
    "result_1.png",
    "result_2.png",
    "result_3.png",
    "result_4.png",
    "result_5.png"
]

user_img = Image.open(f"{INPUT_DIR}/{user_file}")

fig, axes = plt.subplots(len(hair_refs), 3, figsize=(12, 20))

for i in range(len(hair_refs)):
    ref_img = Image.open(f"{INPUT_DIR}/{hair_refs[i]}")
    result_img = Image.open(f"{OUTPUT_DIR}/{result_files[i]}")

    axes[i, 0].imshow(user_img)
    axes[i, 0].set_title("User / Target Face")
    axes[i, 0].axis("off")

    axes[i, 1].imshow(ref_img)
    axes[i, 1].set_title(f"Hair Ref {i+1}")
    axes[i, 1].axis("off")

    axes[i, 2].imshow(result_img)
    axes[i, 2].set_title(f"Result {i+1}")
    axes[i, 2].axis("off")

plt.tight_layout()
plt.show()
```
